## **01-MinMaxScaler**

```python
#01-MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
m = MinMaxScaler()
#m = MinMaxScaler(feature_range=(3,5))
```

```python
#Fitting the Data
scaled1 = m.fit(df[['Marketing Spend']])
```

```python
scaled1
```

```python
df.head(5)
```

```python
#Transform the Data
df['Marketing Spend'] = m.transform(df[['Marketing Spend']])
```
