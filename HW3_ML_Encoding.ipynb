# **1. Label Encoder**

```python
df = pd.read_csv(path)
df.head(5)
```

```python
import pandas as pd
df = pd.read_csv(path)
df.head(5)
```

```python
import numpy as np
from pandas.core.dtypes.common import is_numeric_dtype
for column in df.columns:
  if is_numeric_dtype(df[column]):
    continue
  else:
    df[column]=le.fit_transform(df[column])
```

# **2. One Hot Encoder Using Loop**

```python
import pandas as pd
df = pd.read_csv(path)
df.head(5)
```

```python
columns=['region', 'sex', 'smoker']

for col in columns:
  one_hot_en=pd.get_dummies(df[col], drop_first=True) 
  df=pd.concat((df,one_hot_en), axis=1)
```
